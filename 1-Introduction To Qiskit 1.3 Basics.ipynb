```python
import numpy as np

psi_0 = qi.Statevector([1,1,0,0]/np.sqrt(2))

psi_0.draw('latex')
```

```python
qc3 = QuantumCircuit(2)
qc3.initialize(psi_0,[0,1])
qc3.cx(0,1)

qc3.draw('mpl')
```

```python
psi_1 = psi_0.evolve(qc3)
psi_1.draw('latex')
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
ghz = QuantumCircuit(3)

ghz.h(0)
ghz.save_statevector()
ghz.cx(0,1)

ghz.cx(1,2)

ghz.draw('mpl')
```

```python
from qiskit import transpile

simulator_aer = AerSimulator()

ghz_aer = transpile(ghz,simulator_aer)

results = simulator_aer.run(ghz_aer).result()
statevector = results.get_statevector()

statevector.draw('latex')
```
